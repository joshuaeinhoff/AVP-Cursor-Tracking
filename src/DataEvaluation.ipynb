## Gaze Pattern

```python
df_392892_c0.head()
```

```python
v_x = df_392892_c0["xAvatar"] - df_392892_c0["xCursor"]
v_y = df_392892_c0["yAvatar"] - df_392892_c0["yCursor"]
df_392892_c0["gazeDist"] = np.hypot(v_x, v_y)
v_x1 = df_639274_c0["xAvatar"] - df_639274_c0["xCursor"]
v_y1 = df_639274_c0["yAvatar"] - df_639274_c0["yCursor"]
v_x2 = df_742853_c0["xAvatar"] - df_742853_c0["xCursor"]
v_y2 = df_742853_c0["yAvatar"] - df_742853_c0["yCursor"]


face_x, face_y = 0.0, 0.0

fig, c0 = plt.subplots()
c0.scatter([face_x], [face_y], s=60, label="Face (anchor)")
c0.scatter(face_x + v_x, face_y + v_y, s=12, alpha=0.2, label="User A")
c0.scatter(face_x + v_x1, face_y + v_y1, s=12, alpha=0.2, label="User B")
c0.scatter(face_x + v_x2, face_y + v_y2, s=12, alpha=0.2, label="User C")
c0.set_aspect("equal", adjustable="box")
c0.set_xlabel("X (face-centered)")
c0.set_ylabel("Y (face-centered)")
c0.set_title("Gaze relative to face for case 0")
c0.legend()
```

```python
v_x3 = df_392892_c1["xAvatar"] - df_392892_c1["xCursor"]
v_y3 = df_392892_c1["yAvatar"] - df_392892_c1["yCursor"]
v_x4 = df_639274_c1["xAvatar"] - df_639274_c1["xCursor"]
v_y4 = df_639274_c1["yAvatar"] - df_639274_c1["yCursor"]
v_x5 = df_742853_c1["xAvatar"] - df_742853_c1["xCursor"]
v_y5 = df_742853_c1["yAvatar"] - df_742853_c1["yCursor"]

fig, c1 = plt.subplots()
c1.scatter([face_x], [face_y], s=60, label="Face (anchor)")
c1.scatter(face_x + v_x3, face_y + v_y3, s=12, alpha=0.2, label="User A")
c1.scatter(face_x + v_x4, face_y + v_y4, s=12, alpha=0.2, label="User B")
c1.scatter(face_x + v_x5, face_y + v_y5, s=12, alpha=0.2, label="User C")
c1.set_aspect("equal", adjustable="box")
c1.set_xlabel("X (face-centered)")
c1.set_ylabel("Y (face-centered)")
c1.set_title("Gaze relative to face for case 1")
c1.legend()
```

```python
xA, yA = v_x.to_numpy(), v_y.to_numpy() 
xB, yB = v_x1.to_numpy(), v_y1.to_numpy() 
xC, yC = v_x2.to_numpy(), v_y2.to_numpy() 
xA1, yA1 = v_x3.to_numpy(), v_y3.to_numpy() 
xB1, yB1 = v_x4.to_numpy(), v_y4.to_numpy() 
xC1, yC1 = v_x5.to_numpy(), v_y5.to_numpy()

users = {
    "UserA_case0": (xA,  yA),
    "UserB_case0": (xB,  yB),
    "UserC_case0": (xC,  yC),
    "UserA_case1": (xA1, yA1),
    "UserB_case1": (xB1, yB1),
    "UserC_case1": (xC1, yC1),
}

rect_px = {
    "UserA_case0": (100, 120), 
    "UserB_case0": (60,  80),
    "UserC_case0": (50,  60),
    "UserA_case1": (60,  80),
    "UserB_case1": (100, 120),
    "UserC_case1": (50,  60),
}

items = list(users.items())
n, cols = len(items), 3
rows = math.ceil(n / cols)

fig, axs = plt.subplots(2,3,constrained_layout=True, dpi=300, figsize=(12, 6))
axs = np.atleast_1d(axs).ravel()


xlim = (-500, 500)
ylim = (-1000, 1000)

for ax in axs:
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_aspect("equal", adjustable="box")
    ax.set_box_aspect(1)
    ax.axhline(0, lw=0.6, alpha=0.4)
    ax.axvline(0, lw=0.6, alpha=0.4)
    ax.invert_xaxis()

for i, (name, (x, y)) in enumerate(items):
    ax = axs[i]
    x = np.asarray(x); y = np.asarray(y)


    A, B = rect_px[name]
    rect = Rectangle((-A/2, -B/2), A, B, fill=False, ls='--', lw=1.5)
    ax.add_patch(rect)

    
    inside = (np.abs(x) <= A/2) & (np.abs(y) <= B/2)
    n_in = int(inside.sum()); n_out = int(len(x) - n_in)

    ax.scatter(x, y, s=12, alpha=0.25, edgecolors='none', label=name)
    ax.scatter([0], [0], s=5, zorder=3, label="Face")

    ax.set_title(f"{name}\ninside: {n_in} | outside: {n_out}")
    ax.set_xlabel("X (pixels)")
    if i % cols == 0:
        ax.set_ylabel("Y (pixels)")

for j in range(i+1, len(axs)):
    axs[j].axis('off')

plt.show()
```
